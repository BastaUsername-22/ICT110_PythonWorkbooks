# EXERCISE 1

Investigate on the difference between `is` and `==` given the following variables.
```python
a = 2
b = 2.0
c = a
```

````python
a = 2
b = 2.0
c = a

print(a is b)
print(a is c)
print(b is c)
# 'is' checks identity variable and memory-wise. It only returns true when two variables are correlated (e.g c = a).
print(a == b)
print(a == c)
print(b == c)
# '==' compares value of variables if they have equal value.
````

# CHALLENGE

Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

    Enter a number: 4
    Enter a number: 5
    Enter a number: bad data
    Invalid input
    Enter a number: 7
    Enter a number: done
    16 3 5.333333333333333

````python
total = 0.0
count = 0.0
while(True):
  sample = input("Enter a number: ")

  try:
    float(sample)
  except:
    if(sample == 'done'):
      break
    else:
      print("Invalid Input")
  else:
    number = float(sample)
    total = total + number
    count += 1

if(total != 0):
  print(f"{total} {int(count)} {total/count}")
````
