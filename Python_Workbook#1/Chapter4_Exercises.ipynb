# CHALLENGE

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score             Grade
    [ 97, 100]        1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

    Run the program repeatedly to test the various different values for input.

```python
def computeGrade(g):
  if g >= 0 and g <= 74:
    return "3.0"
  elif g > 74 and g <= 77:
    return "2.5"
  elif g > 77 and g <= 81:
    return "2.25"
  elif g > 81 and g <= 85:
    return "2.0"
  elif g > 85 and g <= 89:
    return "1.75"
  elif g > 89 and g <= 93:
    return "1.5"
  elif g > 93 and g <= 97:
    return "1.25"
  elif g > 97 and g <= 100:
    return "1.0"

score = int(input("Enter score: "))
print(f"Grade is {computeGrade(score)}")
```
