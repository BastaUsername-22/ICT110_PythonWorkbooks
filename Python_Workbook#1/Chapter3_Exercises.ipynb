In the code below, which will never print regardless of the value for x? Edit the code to correct the error.

    if x > 0 :
        print('Above 0')
    elif x >  10 :
        print('Above 10')
    elif x > 50 :
        print('Above 50')
    elif x >  20 :
        print('Above 20')
    else :
        print('Something else')

```python
# Answer: All elif statements will never print since any number greater than 0 will satisfy the first if statement before other statements could evaluate it.

# Revised code:
x = int(input("Enter value for x: "))
if x > 0 and x <= 10 :
    print('Above 0')
elif x > 10 and x <= 20:
    print('Above 10')
elif x > 50:
    print('Above 50')
elif x > 20 and x <= 50:
    print('Above 20')
else :
    print('Something else')
```

# EXERCISE 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0

```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))
if hours <= 40:
  totalRate = hours * rate
elif hours > 40:
  regularRate = 40 * rate
  overtimeRate = ((hours - 40) * rate) * 1.5
  totalRate = regularRate + overtimeRate
print(f"Pay: {totalRate}")
```

# CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score             Grade
    [ 97, 100]        1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

```python
grade = int(input("Enter Grade: "))
if grade >= 0 and grade <= 74:
  print("Grade: 3.0")
elif grade > 74 and grade <= 77:
  print("Grade: 2.5")
elif grade > 77 and grade <= 81:
  print("Grade: 2.25")
elif grade > 81 and grade <= 85:
  print("Grade: 2.0")
elif grade > 85 and grade <= 89:
  print("Grade: 1.75")
elif grade > 89 and grade <= 93:
  print("Grade: 1.5")
elif grade > 93 and grade <= 97:
  print("Grade: 1.25")
elif grade > 97 and grade <= 100:
  print("Grade: 1.0")
```
