# EXERCISE 3

Create a list containing the valid grade values, then using `enumerate` in a `for` loop, print out all **indices** and **items** in the list. 

It should print out:

    0   1.0
    1   1.25
    2   1.5
    3   1.75
    4   2.0
    5   2.25
    6   2.5
    7   3.0

```python
gradeList = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0]

for index, grades in enumerate(gradeList):
    print(index, ' ', grades)
```

# EXERCISE 4

Using Index Slicing, print out the list of grades higher than 2.0

```python
gradeList = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0]

for index, grades in enumerate(gradeList[5:]):
    print(index, ' ', grades)
```

# EXERCISE 5

Using Index Slicing, print out the list of grades better than 2.0.

```python
gradeList = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0]

for index, grades in enumerate(gradeList[:4]):
    print(index, ' ', grades)
```

Modify the code in the subheading [Parsing Lines](#ParsingLines) to display all the time stamps 
of the emails in mboxt-short.txt

```python
fhand = open('mbox-short.txt')
my_list = []
skipState = False
for line in fhand:
    line = line.rstrip()
    if not line.startswith('Date:'): 
        continue
    words = line.split()
    if my_list == []:
        if not skipState:
            my_list.append(line.replace("Date:", "").strip())
            skipState = True
        elif skipState:
            skipState = False
            continue
    elif words[1] not in my_list:
        if not skipState:
            my_list.append(line.replace("Date:", "").strip())
            skipState = True
        elif skipState:
            skipState = False
            continue
print(my_list)

# FOR CHECKING PURPOSES
for item in my_list:
    print(item)
```
